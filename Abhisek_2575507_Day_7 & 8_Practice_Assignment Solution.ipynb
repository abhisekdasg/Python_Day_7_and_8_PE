#   1. Write a Python program to find a target values in a list using linear search with following steps:
    a. Initialize the list to store the input elements.
    b. Initialize found-False.
    C. Enter the item to be searched (match_item).
    d. For each element in the list
            1. if match_item = value
                    a. return match_item's position.
    e. If the match_item is not in the list, display an error message that the item is not found in the list.


```python
def linear_search(input_list, match_item):
    found = False

    for index, value in enumerate(input_list):
        if match_item == value:
            found = True
            return index+1

    return found
```

```python
# Initialize an empty list to store the input elements.
input_list = []
# Input the number of elements you want to add to the list
num_elements = int(input("Enter the number of elements to add to the list: "))

# Input the elements and store them in the list
for i in range(num_elements):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

    print('Your entered data list is: ',input_list)

# Input the item to be searched
match_item = int(input("Enter the item to be searched: "))


# Perform the linear search
result = linear_search(input_list, match_item)

if result is not False:
    print(f"{match_item} found at position {result}")
else:
    print(f"{match_item} is not found in the list.")

```

# 2. Write a Python program to implement binary search to find the target values from the list:
    a. Create a separate function to do binary search.
    b. Get the number of inputs from the user.
    C. Store the inputs individually in a list.
    d. In binary search function at first sort the list in order to start the search from middle of the list.
    e. Compare the middle element to right and left elements to search target element.
    f. If greater, move to right of list or else move to another side of the list.


```python
def binary_search(arr, target):
    arr.sort()  # Sort the list
    
    # prining the list after sorting
    print('After Sorting, Inputed list is: ',arr)

    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid  # if target found, return the index
        elif arr[mid] < target:
            left = mid + 1  # if the target is present in the right half
        else:
            right = mid - 1  # if the target is present in the left half

    return arr -1  # if the target is not found in the list

# Get the number of inputs from the user
num_inputs = int(input("Enter the number of inputs: "))

# Initialize an empty list to store the inputs
input_list = []

# Input the elements individually
for i in range(num_inputs):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Input the targeted element to search
target = int(input("Enter the targeted element you want to search: "))

# printing the elements that a user inputed
print('User Inputed list is: ',input_list)

# Perform binary search
result = binary_search(input_list, target)


if result != -1:
    print(f"{target} found at position {result+1}")
else:
    print(f"{target} is not found in the list.")
```

# 3. Write a Python program for sorting a list of elements using selection sort algorithm:
    a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.
    b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted list.
    C. In the second iteration, find the smallest element in the unsorted list and place it in the correct position by comparing with the element in the sorted list.
    d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position by comparing with the elements in the sorted list.
    e. This process continues till the unsorted list becomes empty.
    f. Display the sorted list.


```python
def selection_sort(arr):
    sorted_list = []
    
    # Loop until the input list becomes empty
    iteration = 0  # Initialize the iteration number
    
    while arr:
        min_index = 0
        
        # Find the index of the minimum element in the unsorted list
        for i in range(1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i
        
        # Move the smallest element from the unsorted list to the sorted list
        smallest_element = arr.pop(min_index)
        sorted_list.append(smallest_element)
        
        # Display the iteration number and the sorted list after each iteration
        iteration += 1
        print(f"Sorted List after Iteration {iteration}: {sorted_list}")
    
    return sorted_list

# Input list of integers from the user using a for loop
input_list = []
num_elements = int(input("Enter the number of elements: "))

for i in range(num_elements):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Display the original list
print("Original List is:", input_list)
print()

# Perform selection sort
sorted_list = selection_sort(input_list.copy())  # Create a copy to keep the original list intact
print()

# Display the final sorted list
print("Final Sorted List is:", sorted_list)
```

# 4. Write a Python program for sorting a list of elements using insertion sort algorithm:
    a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list. 
    b. In the first iteration, take the first element in the unsorted list and insert it in Sorted list.
    C. In the second iteration, take the second element in the given list and compare with the element in the sorted sub list and place it in the correct position.
    d. In the third iteration, take the third element in the given list and compare with the elements in the sorted sub list and place the elements in the correct position.
    e. This process continues until the last element is inserted in the sorted sub list.
    f. Display the sorted elements.


```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Display the current state of the list before insertion
        print(f"Iteration {i}: {arr[:i]} | {arr[i:]}")
        
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        
        # Insert the key into the correct position
        arr[j + 1] = key

# Input list of integers from the user using a for loop
input_list = []
num_elements = int(input("Enter the number of elements: "))

for i in range(num_elements):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Display the original list
print("Original List:", input_list)

print()
# Perform insertion sort
insertion_sort(input_list)
print()

# Display the sorted list
print("Sorted List:", input_list)
```

# 5. Write a Python program that performs merge sort on a list of numbers:
    a. Divide: If the given array has zero or one element, return 1. 
    Otherwise
        ii. Divide the input list in to two halves each containing half of the elements.
            i.e. left half and right half.
    b. Conquer: Recursively sort the two lists (left half and right half).
            a. Call the merge sort on left half.
            b. Call the merge sort on right half.
    C. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.


```python
# Function to perform merge sort
def merge_sort(arr):
    # Divide: If the given array has zero or one element, return it directly.
    if len(arr) <= 1:
        return arr

    # Divide the input list into two halves (left half and right half)
    middle = len(arr) // 2
    left_half = arr[:middle]
    right_half = arr[middle:]

    # Display the two lists after the divide step
    print("Divided Lists:", left_half, right_half)
    
    # Conquer: Recursively sort the two lists (left half and right half)
    left_half = merge_sort(left_half)  # Call merge sort on the left half
    right_half = merge_sort(right_half)  # Call merge sort on the right half

    # Display the two lists after the conquer step
    print("Conquered Lists:", left_half, right_half)

    # Combine: Merge the two sorted halves into a single sorted sequence
    merged_arr = merge(left_half, right_half)
    return merged_arr

# Function to merge two sorted lists
def merge(left, right):
    merged = []
    left_index, right_index = 0, 0

    while left_index < len(left) and right_index < len(right):
        # Compare elements from the left and right lists and merge them
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    # Append any remaining elements from the left and right lists
    merged.extend(left[left_index:])
    merged.extend(right[right_index:])
    
    return merged

# Input list of numbers from the user using a for loop
input_list = []
num_elements = int(input("Enter the number of elements: "))

for i in range(num_elements):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Display the original list
print("Original List:", input_list)

# Perform merge sort
sorted_list = merge_sort(input_list.copy())  # Create a copy to keep the original list intact

# Display the sorted list
print("Sorted List:", sorted_list)
```

# 6. Write a Python script to perform the following operations on a singly linked list
    a. Create a list
    b. Find the smallest element from the list
    C. Insert an element if it is not a duplicate element
    d. Display the elements in reverse order

```python
# Define a Node class to represent elements in the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Define a SinglyLinkedList class
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def create_list(self, elements):
        # Function to create a linked list from a list of elements
        for element in elements:
            self.append(element)
   

    def append(self, data):
        # Function to append a new node with the given data at the end of the linked list
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def find_smallest(self):
        # Function to find and return the smallest element in the linked list
        if not self.head:
            return None

        smallest = self.head.data
        current = self.head.next

        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next

        return smallest

    def insert_if_not_duplicate(self, data):
        # Function to insert a new element with the given data if it is not a duplicate

        if not self.head:
            self.head = Node(data)
            return True

        current = self.head
        prev = None

        while current:
            if current.data == data:
                return False  # Data already exists, don't insert

            prev = current
            current = current.next

        # Data is not a duplicate, insert it at the end
        prev.next = Node(data)
        return True
    
    def display_list(self):
        if not self.head:
            print("The list is empty.")
            return

        current = self.head
        nodes = []
        while current:
            nodes.append(str(current.data))
            current = current.next
        print("[" + "----->".join(nodes) + "]")


    def display_reverse(self):
    # Function to display the elements of the linked list in reverse order
        stack = []
        current = self.head

        while current:
            stack.append(str(current.data))
            current = current.next

        print("[" + "----->".join(stack[::-1]) + "]")

# Create an empty singly linked list
linked_list = SinglyLinkedList()

# Get the number of inputs from the user
num_inputs = int(input("Enter the total number elements you want to insert: "))

# cearting an empty list to store the input data
elements = []
# Take user input for elements directly as integers
try:
    for i in range(num_inputs):
            element = int(input(f"Enter integer {i + 1} to adding to the list (Enter non-integer to stop): "))
            elements.append(element)
except ValueError as e:
        print(f'Erorr!!!{e}')
        

# Add elements to the linked list
linked_list.create_list(elements)

# printing list elements
print()
print("Printing the list that you have entered: ")
linked_list.display_list()


# Find the smallest element
smallest = linked_list.find_smallest()
if smallest is not None:
    print(f"\nThe smallest element in the list is: {smallest}")
else:
    print("The list is empty.")

print()
# Take user input for an element to insert
element_to_insert = int(input("Enter an element (integer) to insert (or Enter non-integer to skip insertion): "))
if linked_list.insert_if_not_duplicate(element_to_insert):
    print(f"{element_to_insert} was inserted into the list.")
else:
    print(f"{element_to_insert} is already in the list or insertion was skipped.")

print()
print("Printing the list after your insertion: ")
linked_list.display_list()
print()

# Display the elements in reverse order
print("Elements in reverse order:")
linked_list.display_reverse()
```

# 7.Write a python program to implement the various operations for Stack ADT
    i.) Push ii.) Pop iii.) Display.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            print("Stack is empty. Cannot pop an element.")
            return None

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        if not self.is_empty():
            print("Stack elements:")
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty.")

# Create a stack object
stack = Stack()

while True:
    print("\n**Stack Operations:**")
    print("1. Push/Insert")
    print("2. Pop/Delete")
    print("3. Display")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        item = input("Enter the element to Push/Insert: ")
        stack.push(item)
        print(f"{item} Pushed/Inserted into the stack.")
    elif choice == 2:
        popped_item = stack.pop()
        if popped_item is not None:
            print(f"Popped element: {popped_item}")
    elif choice == 3:
        stack.display()
    elif choice == 4:
        break
    else:
        print("Wrong Input.. Please try again with right input.")
```

# 8. Write a python script to implement the various operations for Queue ADT
    i.) Insert ii.) Delete iii.) Display.

```python
class Queue:
    def __init__(self):
        self.items = []

    def insert(self, item):
        self.items.append(item)

    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            print("Queue is empty. Cannot delete an element.")
            return None

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        if not self.is_empty():
            print("Queue elements:")
            for item in self.items:
                print(item)
        else:
            print("Queue is empty.")

# Create a queue object
queue = Queue()

while True:
    print("\n***Queue Operations:***")
    print("1. Insert")
    print("2. Delete")
    print("3. Display")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        item = input("Enter the element to insert: ")
        queue.insert(item)
        print(f"{item} inserted into the queue.")
    elif choice == 2:
        deleted_item = queue.delete()
        if deleted_item is not None:
            print(f"Deleted element: {deleted_item}")
    elif choice == 3:
        queue.display()
    elif choice == 4:
        break
    else:
        print("Invalid choice. Please try again.")
```

# 9. Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack
        a. A/B^C+D*E-F*G
        b. (B^2-4*A*C)^(1/2) (100)

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(expression):
    precedence = {
        '^': 3,
        '*': 2,
        '/': 2,
        '+': 1,
        '-': 1,
        '(': 0
    }
    
    output = []
    stack = Stack()

    for char in expression:
        if char.isalnum():
            output.append(char)
        elif char == ' ':
            continue  # Skip spaces
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while not stack.is_empty() and stack.peek() != '(':
                output.append(stack.pop())
            if not stack.is_empty() and stack.peek() == '(':
                stack.pop()
        else:
            while not stack.is_empty() and precedence[char] <= precedence.get(stack.peek(), 0):
                output.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        output.append(stack.pop())

    return ''.join(output)

# Test cases
infix_expression_one = "A/B^C+D*E-F*G"
infix_expression_two = "(B^2-4*A*C)^(1/2) (100)"

postfix_one = infix_to_postfix(infix_expression_one)
postfix_two = infix_to_postfix(infix_expression_two)

print("Infix Expression one:", infix_expression_one)
print("Postfix Expression one:", postfix_one)
print()
print("Infix Expression two:", infix_expression_two)
print("Postfix Expression two:", postfix_two)
```
